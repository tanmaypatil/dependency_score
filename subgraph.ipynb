```python
import graphviz
```

```python
p = graphviz.Graph('parent')  
```

```python
display(p)
```

```python
with p.subgraph(name='cluster', node_attr={'shape': 'box'}) as c:
   c.attr('graph',color='red')
   c.attr(label='process #1')
   c.node_attr.update(style='filled', color='lightgrey')
   c.edge('foo', 'bar')
```

```python

display(c)
```
