#Ways of visualisation of dependency scores.  
Scores of Product increment 1 are present in scores_pi1 array [2, 2.1, 4, 0.5, 3.1]     
Scores of Product increment 1 are present in scores_pi2 array [3.9, 4, 4.1, 4.6, 4.8, 1.0]
* boxplot
* violin plot


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dependency scores for each product increment
scores_pi1 = [2, 2.1, 4, 0.5, 3.1]
scores_pi2 = [3.9, 4, 4.1, 4.6, 4.8, 1.0]
```

```python
# Creating a DataFrame
df = pd.DataFrame({
    'Dependency Score': scores_pi1 + scores_pi2,
    'Product Increment': ['PI1'] * len(scores_pi1) + ['PI2'] * len(scores_pi2)
})
```

```python
df
```

```python
# Plotting using Matplotlib
plt.figure(figsize=(8, 6))
plt.boxplot([scores_pi1, scores_pi2], labels=['Product Increment 1', 'Product Increment 2'])
plt.title('Dependency Scores of Epics in Product Increments 1 vs 2')
plt.ylabel('Dependency Score')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Product Increment', y='Dependency Score', data=df)
plt.title('Dependency Scores of Epics in Product Increments 1 vs 2')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Product Increment', y='Dependency Score', data=df)
plt.title('Dependency Scores of Epics in Product Increments 1 vs 2')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.swarmplot(x='Product Increment', y='Dependency Score', data=df)
plt.title('Swarm Plot of Dependency Scores')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.ecdfplot(data=df, x='Dependency Score', hue='Product Increment')
plt.title('Cumulative Distribution of Dependency Scores')
plt.xlabel('Dependency Score')
plt.ylabel('Proportion of Epics')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(df[df['Product Increment'] == 'PI1']['Dependency Score'], label='Product Increment 1', fill=True)
sns.kdeplot(df[df['Product Increment'] == 'PI2']['Dependency Score'], label='Product Increment 2', fill=True)
plt.title('Density Plot of Dependency Scores')
plt.xlabel('Dependency Score')
plt.ylabel('Density')
plt.legend()
plt.show()
```
