## Multiple subgraphs in a graph  
Create 2 subgraphs . one under p1 and another under p2  
p1 has 2 dependent nodes under it c11 and c21   
p2 has 2 dependent nodes under it c21 and c22
add both parent nodes in set p1 ,p2 . Our final api will pass set of parent nodes 
draw_allsubgraphs function will return the graph

```python
p1 = Node("P2G-100","FUNC")
c11 = Node("P2G-1000","DEVOPS",15)
c21 = Node("P2G-1001","INFRA",16)
p1.add_dependency(c11)
p1.add_dependency(c21)

p2 = Node("P2G-200","FUNC")
c21 = Node("P2G-2000","DEVOPS",25)
c22 = Node("P2G-2001","INFRA",26)
p2.add_dependency(c21)
p2.add_dependency(c22)

n = set()
n.add(p1)
n.add(p2)
g = draw_allsubgraphs(n)
display(g)
```
