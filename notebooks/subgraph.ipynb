## Multiple subgraphs in a graph  
Create 2 subgraphs . one under p1 and another under p2  
p1 has 2 dependent nodes under it c11 and c21   
p2 has 2 dependent nodes under it c21 and c22
add both parent nodes in set p1 ,p2 . Our final api will pass set of parent nodes 
draw_allsubgraphs function will return the graph

```python
p1 = Node("P2G-100","FUNC")
c11 = Node("P2G-1000","DEVOPS",15)
c21 = Node("P2G-1001","INFRA",16)
p1.add_dependency(c11)
p1.add_dependency(c21)

p2 = Node("P2G-200","FUNC")
c21 = Node("P2G-2000","DEVOPS",25)
c22 = Node("P2G-2001","INFRA",26)
p2.add_dependency(c21)
p2.add_dependency(c22)

n = set()
n.add(p1)
n.add(p2)
g = draw_allsubgraphs(n)
display(g)
```

## Integrated with node hierarchy 
There are 2 functional requirements P2G-100 and P2G-101 with there respective dependencies .
2 seperate subgraphs are added . 


```python
anodes = AllNodes()
n = anodes.create_node("P2G-100","FUNC")
d = anodes.create_node("GDB-100","DEVOPS",2)
d1 = anodes.create_node("PCM-100","INFRA",3)
anodes.add_dependency(n,d)
anodes.add_dependency(n,d1)
# 2nd parent node
n1 = anodes.create_node("P2G-101","FUNC")
d11 = anodes.create_node("GDB-101","DEVOPS",5)
d12 = anodes.create_node("PCM-102","INFRA",7) 
anodes.add_dependency(n1,d11)
anodes.add_dependency(n1,d12)
_,parents = anodes.get_top_parents()
g= draw_allsubgraphs(parents)
display(g)
```
