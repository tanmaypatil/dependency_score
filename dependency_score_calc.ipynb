```python
def draw_tree1(root, format='svg', rankdir='TB'):
    dot = Digraph(format=format, graph_attr={'rankdir': rankdir})
    c1 = Node("GDB-1000","DEVOPS",5)
    dot.node(name=str(id(c1)), label = "{ id %s | dep_score %.4f }" % (c1.id ,c1.get_dependency_score()), shape='record')
    c2 = Node("PCM-1001","INFRA",5)
    dot.node(name=str(id(c2)), label = "{ id %s | dep_score %.4f }" % (c2.id ,c2.get_dependency_score()), shape='record')
    p = Node("P2G-100","FUNC",5)
    dot.node(name=str(id(p)), label = "{ id %s | dep_score %.4f }" % (p.id ,p.get_dependency_score()), shape='record')
    dot.edge(str(id(p)), str(id(c1)))
    dot.edge(str(id(p)), str(id(c2)))
    p2 = Node("P2G-99","FUNC",5)
    dot.node(name=str(id(p2)), label = "{ id %s | dep_score %.4f }" % (p2.id ,p2.get_dependency_score()), shape='record')
    dot.edge(str(id(p2)),str(id(p)))
    return dot
```

```python
from graphviz import Digraph
from IPython.display import display
dot_1 = draw_tree1(None)
display(dot_1)
```
